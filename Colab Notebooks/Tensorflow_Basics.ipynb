Tensor aggregation

```python
import random
```

```python
random.seed(108)
a = random.sample(range(1,100), 50)
a = tf.constant(a)
a
```

```python
print("The samllest number:", tf.reduce_min(a))
print("The largest number:", tf.reduce_max(a))
print("The mean of the numbers:", tf.reduce_mean(a))
print("The sum of the numbers:", tf.reduce_sum(a))
```

```python
print("Index of the samllest number", tf.argmin(a))
print("Index of the largest number", tf.argmax(a))
```
