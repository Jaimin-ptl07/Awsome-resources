```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [20,23,25,28,30,37,40,43,46] # Hours of study
y = [45,51,55,61,65,79,85,91,97] # Test scores
np.shape(X)
```

```python
plt.plot(X,y)
plt.title("Exam performance graph")
plt.xlabel("Hours of study")
plt.ylabel("Test scores")
plt.show()
```

```python
study_model = Sequential([Dense(units=1,input_shape = [1])])
study_model.compile(optimizer = "adam", loss = 'mean_squared_error')
X = np.array(X)
y = np.array(y)
```

```python
history = study_model.fit(X,y,epochs = 1000)
```

```python
n=40
n = np.array([n])
result = study_model.predict([n])[0][0]
```
