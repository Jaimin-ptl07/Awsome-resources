```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('/content/Salary Data.csv')
df.head()
```

```python
X = df.drop("Salary", axis='columns')
y = df["Salary"]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, random_state=108, test_size=0.2)
```

```python
X_test.shape
```

```python
#base model
model = Sequential([Dense(1, activation='linear', input_shape=[len(X_train.columns)])])
model.compile(loss=tf.keras.losses.mae, optimizer = tf.keras.optimizers.SGD(), metrics=['mae'])
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
# increased number of eopchs
model_1 = Sequential([Dense(1, activation='linear', input_shape=[len(X_train.columns)])])
model_1.compile(loss=tf.keras.losses.mae, optimizer = tf.keras.optimizers.SGD(), metrics=['mae'])
history = model_1.fit(X_train,y_train, epochs=500)
```

```python
def visualize_model(history, ymin=None, ymax=None):
  print(history.history.keys())
  plt.plot(history.history['loss'])
  plt.title('Model loss')
  plt.ylabel('loss')
  plt.xlabel('number of epochs')
  plt.ylim([ymin,ymax])
  plt.legend(['loss plot'])
  plt.show()
```

```python
visualize_model(history)
```

```python
#we have added one more layer of 64 neurons as model was underfitting
tf.random.set_seed(10)

model_2 = Sequential([Dense(units=64,activation='relu',input_shape=[len(X_train.columns)]),
                      Dense(units=1)
                      ])
model_2.compile(loss='mae', optimizer='SGD', metrics=['mae'])

history_2 = model_2.fit(X_train,y_train,epochs=500)
```

```python
visualize_model(history_2, ymin=0,ymax=10000)
```

```python
#add one more layer of 64 neurons is making model overfitting
tf.random.set_seed(10)

model_3 = Sequential([
   Dense(units=64,activation='relu',input_shape=[len(X_train.columns)]),
   Dense(units=64,activation='relu'),
   Dense(units=1)
])

model_3.compile(loss='mae',optimizer='SGD',metrics=['mae'])

history_3 = model_3.fit(X_train,y_train,epochs=500)
```

normalization

```python
X.describe
```

```python
scaler = MinMaxScaler()

X_norm = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
X_norm.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_norm, y, random_state=108, test_size=0.2)
```

```python
#model with normalization of data with early stopping
model_4 = Sequential([
    Dense(units=64,activation='relu',input_shape=[len(X_train.columns)]),
    Dense(units=1)
])

model_4.compile(loss='mae', optimizer='SGD',metrics=['mae'])

history_4 = model_4.fit(X_train, y_train, epochs=1000, callbacks=tf.keras.callbacks.EarlyStopping(patience=10))
```

```python
visualize_model(history_4)
```

```python
#Adam optimizer
model_5 = Sequential([
    Dense(units=64,activation='relu',input_shape=[len(X_train.columns)]),
    Dense(units=1)
])

model_5.compile(loss='mae',optimizer='Adam',metrics=['mae'])

model_5.fit(X_train,y_train,epochs=1000,callbacks=[tf.keras.callbacks.EarlyStopping(patience=10)])
```

```python
def eval_testing(model):
  return model.evaluate(X_test,y_test)

models=[model,model_1,model_2,model_3,model_4,model_5]

for x in models:
  eval_testing(x)
```
